```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

penguins = sns.load_dataset('penguins')
```

```python
penguins = pd.read_csv("https://raw.githubusercontent.com/allisonhorst/palmerpenguins/refs/heads/main/inst/extdata/penguins.csv")
```

```python
fig, axes = plt.subplots(1, 4, figsize=(20, 5))

# plot A
sns.histplot(data=penguins,
             x='body_mass_g',
             bins=20, alpha=0.6,
             hue='species',
             palette='magma',
             kde=True,
             ax=axes[0])  #subplot 1
axes[0].set_title('Chart A')
axes[0].set_xlabel('Body Mass (g)')
axes[0].set_ylabel('Frequency')

# plot B
sns.boxplot(data=penguins,
            x='body_mass_g',
            hue='species',
            palette='magma',
            ax=axes[1])  #subplot 2
axes[1].set_title('Chart B')
axes[1].set_xlabel('Body Mass (g)')

# plot C
sns.violinplot(data=penguins,
               y='species',
               x='body_mass_g',
               hue='species',
               palette='magma',
               density_norm ='count',
               ax=axes[2])  #subplot 3
axes[2].set_title('Chart C')
axes[2].set_xlabel('Body Mass (g)')
axes[2].set_ylabel(None)

# plot D
sns.barplot(data=penguins,
            y='species',
            x='body_mass_g',
            hue='species',
            palette='magma',
            ax=axes[3])
axes[3].set_title('Chart D')
axes[3].set_xlabel('Body Mass (g)')
axes[3].set_ylabel(None)
```

#Visualizations


```python
penguins['body_mass_g'].plot(kind='hist', bins=20,color='indigo', edgecolor='black')
```

```python
penguins['body_mass_g'].describe()
```

```python
penguins['species'].value_counts()
```

```python
sns.countplot(data=penguins,
   x='species',
   color='tab:olive')

plt.title('Barchart: Species Count')
plt.xlabel(None)
plt.ylabel('Count')
```

```python
sns.scatterplot(data=penguins,
 		    x='bill_length_mm',
    y='bill_depth_mm',
    hue = 'species',
    palette='viridis')
plt.title('Bill Dimensions by Species')
plt.xlabel('Length (mm)')
plt.ylabel('Depth (mm)')
```

```python
sns.lmplot(data=penguins,
          x='bill_length_mm',
          y='bill_depth_mm',
          hue="species",
          palette='viridis',
          legend=None, height=6)
plt.legend(loc='lower left', ncol=1)
```

```python
pivot_data = penguins.pivot_table(index='species',
columns='sex',
aggfunc='size',
fill_value=0)

pivot_data.plot(kind='bar', stacked=True,
    color= ['green', 'tab:olive'])
```

```python
sns.boxplot(data = penguins, x = 'island',y ='body_mass_g',
  hue = 'species', palette = 'crest', linewidth=0.3)
```

```python
num_penguins = penguins.select_dtypes(include = ['float64', 'int64'])
corr_matrix = num_penguins.corr()
sns.heatmap(corr_matrix, annot=True, cmap='magma', fmt='.2f', cbar=True)
```

```python
sns.pairplot(penguins, hue='species',palette='magma', dropna=True)
```
